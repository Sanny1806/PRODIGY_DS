```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load the dataset with corrected file path
df = pd.read_csv('D:\\Data\\countries-table.csv')

# Display the first few rows of the dataset
print(df.head())
```

```python
# Create bar chart
plt.figure(figsize=(10,6))
plt.bar(df['country'], df['pop2023'], color='skyblue')
plt.xlabel('Countries')
plt.ylabel('Population in 2023')
plt.title('Population Distribution Across Top 14 Countries in 2023')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

# Show plot
plt.show()
```
