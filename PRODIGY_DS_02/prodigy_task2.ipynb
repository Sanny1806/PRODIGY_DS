```python
import pandas as pd
import numpy as np
```

```python
# Load the dataset with corrected file path
titanic_df = pd.read_csv('D:\\Data\\titanic.csv')

# Display the first few rows of the dataset
print(titanic_df.head())
```

```python
# Check for missing values
print(titanic_df.isnull().sum())

# Fill missing values for 'Age' with the median age
median_age = titanic_df['Age'].median()
titanic_df['Age'].fillna(median_age, inplace=True)

# Fill missing values for 'Embarked' with the mode
mode_embarked = titanic_df['Embarked'].mode()[0]
titanic_df['Embarked'].fillna(mode_embarked, inplace=True)

# Drop 'Cabin' column as it has too many missing values
titanic_df.drop('Cabin', axis=1, inplace=True)

# Verify missing values have been handled
print(titanic_df.isnull().sum())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Distribution of passengers by sex
sns.countplot(x='Sex', data=titanic_df)
plt.title('Passenger Distribution by Sex')
plt.show()

# Survival rate by sex
sns.countplot(x='Sex', hue='Survived', data=titanic_df)
plt.title('Survival Count by Sex')
plt.show()

# Distribution of passengers by class
sns.countplot(x='Pclass', data=titanic_df)
plt.title('Passenger Distribution by Class')
plt.show()

# Survival rate by class
sns.countplot(x='Pclass', hue='Survived', data=titanic_df)
plt.title('Survival Count by Class')
plt.show()

# Distribution of passengers by age
sns.histplot(data=titanic_df, x='Age', bins=20, kde=True)
plt.title('Passenger Distribution by Age')
plt.xlabel('Age')
plt.show()

# Survival rate by age
sns.histplot(data=titanic_df, x='Age', hue='Survived', bins=20, kde=True)
plt.title('Survival Count by Age')
plt.xlabel('Age')
plt.show()
```
